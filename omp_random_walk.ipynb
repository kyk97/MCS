```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
f = open('TS.txt', 'r')
data = f.read()
f.close()
nums = list(map(float,data.split()))
x = np.zeros(0)
y = np.zeros(0)
for i in range(len(nums)): 
    if i % 4 == 3: 
        y = np.append(y, nums[i])
    else: 
        if i % 4 == 2:
            x = np.append(x, nums[i])
plt.figure(figsize=(10, 8))
plt.xlabel('N')
plt.ylabel('T, s')
plt.grid(True)
plt.title('T(N) for a = 0, b = 100, x = 50, p = 0.49, P = 4')
plt.plot(y, x)
plt.show()
```

```python
f = open('TS.txt', 'r')
data = f.read()
nums = list(map(float,data.split()))
f.close()
x = np.zeros(0)
y = np.zeros(0)
for i in range(len(nums)): 
    if i % 4 == 3: 
        y = np.append(y, nums[i])
    else: 
        if i % 4 == 2:
            x = np.append(x, nums[i - 2]/nums[i])
plt.figure(figsize=(10, 8))
plt.xlabel('N')
plt.ylabel('S')
plt.grid(True)
plt.title('S(N) for a = 0, b = 100, x = 50, p = 0.49, P = 4')
plt.plot(y, x)
plt.show()
```

```python
f = open('TSE.txt', 'r')
data = f.read()
nums = list(map(float,data.split()))
f.close()
x = np.zeros(0)
y = np.zeros(0)
for i in range(len(nums)): 
    if i % 2 == 1: 
        y = np.append(y, nums[i])
    else :
        x = np.append(x, nums[i])
plt.figure(figsize=(10, 8))
plt.xlabel('P')
plt.ylabel('T')
plt.grid(True)
plt.title('T(P) for a = 0, b = 100, x = 50, p = 0.49, N = 100000')
plt.plot(y, x)
plt.plot(y, x, 'ro')
plt.show()
```

```python
f = open('TSE.txt', 'r')
data = f.read()
nums = list(map(float,data.split()))
f.close()
x = np.zeros(0)
y = np.zeros(0)
for i in range(len(nums)): 
    if i % 2 == 1: 
        y = np.append(y, nums[i])
    else :
        x = np.append(x, nums[0] / nums[i] / ((i / 2) + 1))
plt.figure(figsize=(10, 8))
plt.xlabel('P')
plt.ylabel('E')
plt.grid(True)
plt.title('E(P) for a = 0, b = 100, x = 50, p = 0.49, N = 100000')
plt.plot(y, x)
plt.plot(y, x, 'ro')
plt.show()
```
